```python
import pandas as pd
import re
import datetime as dt
```

```python
with open('dataset/WhatsApp Chat with BCT 2081 AB.txt','r',encoding='utf-8') as f:
    data=f.read()
```

```python
print(data)
```

```python
pattern=r"\d{1,2}/\d{1,2}/\d{2,4},\s\d{2}:\d{2}\s-\s"
```

```python
message=re.split(pattern,data)[1:]
print(message)
```

```python
date=re.findall(pattern,data)
print(date)
```

```python
msg_df=pd.DataFrame({'Date':date,'Message':message})
```

```python
msg_df.head()
```

```python
msg_df['Date'] = pd.to_datetime(
    msg_df['Date'],
    format='%d/%m/%y, %H:%M - ',
    errors='coerce'
)
```

```python
msg_df.head()
```

```python
users=[]
messages=[]
for  message_n in msg_df['Message']:
    entry=re.split(r'([\w\W]+?):\s',message_n)
    if entry[1:]:
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('group_notification')
        messages.append(entry[0])
msg_df['Users']=users
msg_df['Message']=messages
```

```python
msg_df.head()
```

```python
msg_df['Year']=msg_df['Date'].dt.year
```

```python
msg_df.head()
```

```python
msg_df['Month']=msg_df['Date'].dt.month_name()
```

```python
msg_df['Day']=msg_df['Date'].dt.day
```

```python
msg_df['Hour']=msg_df['Date'].dt.hour
```

```python
msg_df['Minute']=msg_df['Date'].dt.minute
```

```python
msg_df['Day_Week']=msg_df['Date'].dt.day_name()
```
